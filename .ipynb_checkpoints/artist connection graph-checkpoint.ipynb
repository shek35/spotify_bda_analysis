## find the list of tracks performed by collaboration(preprocessing)

```python
def create_collaboration():
    artist_colab=[]
    for i in range(len(data)):
        if len(data['artists'][i].split(','))>1:
            temp = list();
            temp.append(data['artists'][i])
            temp.append(data['popularity'][i])
            artist_colab.append(temp)
    artist_colab_cleaned=[]
    for item in artist_colab:
        temp = list();
        temp.append(item[0].replace('[','').replace(']','').replace("'",'').strip().split(','))
        temp.append(item[1])
        artist_colab_cleaned.append(temp)
    collab=[]
    for item in artist_colab_cleaned:
        for i in range(1,len(item)):
            temp=[]
            temp.append([item[0][0],item[0][i].strip(' ')])
            temp.append(item[1])
            collab.append(temp)
    collab = np.array(collab)
    collab = collab.reshape(collab.shape[0],2)
    return collab
```

```python
collab = create_collaboration()
collab[:100]
```

## To create graph with frequency of collaboration as weights

```python
weights = dict();
for i in collab:
    temp = str(i[0][0]) + ',' + str(i[0][1])
    if temp in list(weights.keys()):
        weights[temp] += 1
    else:
        weights[temp] = 1
```

```python
weights
```

```python
temp = list(weights.keys())[:20]
temp_weights = list(weights.values())[:20]
```

## To create graph with average popularity as weights

```python
collab[0][1]
```

```python
collab1 = list();
for i in collab:
    flag = 0
    for j in range(len(collab1)):
        if collab1[j][0] == i[0]:
            collab1[j][1] += i[1]
            collab1[j][2] += 1
            flag = 1
    if flag == 0:
        temp = list();
        temp.append(i[0])
        temp.append(i[1])
        temp.append(1)
        collab1.append(temp)
collab1    
weights1 = list();
for i in range(len(collab1)):
    temp = list();
    score = collab1[i][1]/collab1[i][2]
    temp.append(collab1[i][0])
    temp.append(score)
    weights1.append(temp)
weights1
```
