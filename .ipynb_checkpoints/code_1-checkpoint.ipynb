```python
import pandas as pd
import numpy as np
import nltk
import networkx as nx
import matplotlib.pyplot as plt
import pickle
```

```python
data = pd.read_csv('tracks.csv')
```

```python
data
```

```python
data.columns
```

## find the list of tracks performed by collaboration

```python
artist_colab=[]
for item in data['artists']:
    if len(item.split(','))>1:
        artist_colab.append(item)
artist_colab_cleaned=[]
for item in artist_colab:
    artist_colab_cleaned.append(item.replace('[','').replace(']','').replace("'",'').strip().split(','))
```

```python
artist_colab_cleaned
```

## find the names of all the artists

```python
all_artist=[]
for item in artist_colab_cleaned:
    for i in range(len(item)):
        all_artist.append(item[i])
```

```python
all_artist
```

```python
fdist = nltk.FreqDist(all_artist)
```

```python
unique_artist = list(set(all_artist))
del unique_artist[0]
unique_artist
```

```python
artist_df = pd.DataFrame({'artist':list(fdist.keys()), 'count':list(fdist.values())})
artist_df
```

```python
collab=[]
for item in artist_colab_cleaned:
    for i in range(1,len(item)):
        temp=[]
        temp.append([item[0],item[i]])
        collab.append(temp)
```

```python
edges = collab
unique_edge=[]
for i in range(len(edges)):
    if len(edges[1][0])==2:
        if edges[i][0] not in unique_edge:
            unique_edge.append(edges[i][0])
```

```python
collab = np.array(collab)
collab = collab.reshape(collab.shape[0],2)
print(collab.shape)
```

```python
unique_edge = np.array(unique_edge)
unique_edge = unique_edge.reshape(unique_edge.shape[0],2)
print(unique_edge.shape)
```

```python
unique_edge
```

```python
for i in range(len(unique_edge)):
    unique_edge[i,1] = unique_edge[i,1].strip(' ')
```

```python
unique_edge
```

```python
weights = dict();
for i in collab:
    i = str(i[0]) + ',' + str(i[1])
    if i in list(weights.keys()):
        weights[i] += 1
    else:
        weights[i] = 1
```

```python
weights
```

```python
max_weight = max(list(weights.values()))
print(type(max_weight))
```

```python
weights_temp = weights
```

```python
for i in weights_temp:
    weights[i] = round((weights[i]/max_weight),4)
weights
```

## This part is just for reading and writing

```python
try:
    geeky_file = open('weights', 'wb')
    pickle.dump(weights, geeky_file)
    geeky_file.close()
  
except:
    print("Something went wrong")
```

```python
file_to_read = open("weights", "rb")

loaded_dictionary = dict(pickle.load(file_to_read))

print(loaded_dictionary)
```

```python
weights = loaded_dictionary
```

```python
real_edges = []
for j in list(weights.keys()):
    temp = j.split(',')
    real_edges.append(temp)
```

```python
real_edges[1]
```

```python
for i in range(len(real_edges)):
    real_edges[i].append(weights[list(weights.keys())[i]])
    real_edges[i] = tuple(real_edges[i])
```

```python
G = nx.Graph()
```

```python
G.add_weighted_edges_from(real_edges)
```

```python
G1 = nx.subgraph(G, ['Dick Haymes', 'Benny Goodman Sextet',' Gordon Jenkins'])
```

```python
nx.draw(G1)
plt.show()
```

```python
with open('finale.txt','wb') as fp:
    pickle.dump(real_edges,fp)
```
