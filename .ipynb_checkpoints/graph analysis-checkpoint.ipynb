## Data Preprocessing: find the list of tracks performed by collaboration

```python
def get_collaboration_detail():
    artist_colab=[]
    for i in range(len(data)):
        if len(data['artists'][i].split(','))>1:
            temp = list();
            temp.append(data['artists'][i])
            temp.append(data['popularity'][i])
            artist_colab.append(temp)
    artist_colab_cleaned=[]
    for item in artist_colab:
        temp = list();
        temp.append(item[0].replace('[','').replace(']','').replace("'",'').strip().split(','))
        temp.append(item[1])
        artist_colab_cleaned.append(temp)
    collab=[]
    for item in artist_colab_cleaned:
        for i in range(1,len(item)):
            temp=[]
            temp.append([item[0][0],item[0][i]])
            temp.append(item[1])
            collab.append(temp)
    collab = np.array(collab)
    collab = collab.reshape(collab.shape[0],2)
    return collab
```

```python
collab = get_collaboration_detail()
collab
```

## create data for graph with frequency of collaboration as edge weights

```python
weights = dict();
for i in collab:
    i = str(i[0][0]) + ',' + str(i[0][1]).strip(' ')
    if i in list(weights.keys()):
        weights[i] += 1
    else:
        weights[i] = 1
```

```python
weights
```

### create data for graph with average popularity as edge weights

```python
collab1 = list();
flag = 0
for i in collab:
    for j in range(len(collab1)):
        flag = 0
        if collab1[j][0] == i[0]:
            collab1[j][1] += i[1]
            collab1[j][2] += 1
            flag = 1
    if flag == 0:
        temp = list();
        temp.append(i[0])
        temp.append(i[1])
        temp.append(1)
        collab1.append(temp)
weights1 = list();
for i in range(len(collab1)):
    temp = list();
    artist_name = str(collab1[i][0][0]) + ',' + str(collab1[i][0][1]).strip(' ')
    score = collab1[i][1]/collab1[i][2]
    temp.append(artist_name)
    temp.append(score)
    weights1.append(temp)
weights1     
```

## This part is just for reading and writing

```python
temp = list(weights.keys())[:20]
temp_weights = list(weights.values())[:20]
temp_weights
```

```python
unique = list();
for i in range(len(temp)):
    unique.append(temp[i].split(',')[0])
    unique.append(temp[i].split(',')[1])
unique = list(set(unique))
unique
```

```python
temp
```

```python
net1=Network(height='800px', width='100%',notebook=True)
for i in unique:
    net1.add_node(i)
for i in range(len(temp)):
    net1.add_edge(temp[i].split(',')[0], temp[i].split(',')[1], value = temp_weights[i] , title = temp_weights[i])
```
